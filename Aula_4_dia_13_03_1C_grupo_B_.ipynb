```python
# Calcular a área de um retângulo

# Entrada

comprimento = float(input("digite o primeiro número: "))
altura = float(input("digite o primeiro número: "))

# Processamento

area = comprimento * altura

# saída

print(f"O total da área do retângulo é igual: {area}")
```

```python
# Escrever um programa que faça calculos aritméticos

# Entrada

a= int (input ("digite o valor do primeiro número:"))
b= int (input ("digite o valor do primeiro número:"))

# Processamento

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b

# Saída

print ("|--------------------------CALCULADORA SIMPLES--------------------------|")
print ("|                                                                       |")
print ("|---------------------------------SOMA----------------------------------|")
print ("|                                                                       |")
print (f"o resultado da soma de {a} mais {b} é igual a: {soma}")
print ("|                                                                       |")
print ("|------------------------------subtração--------------------------------|")
print ("|                                                                       |")
print (f"o resultado da sutração de {a} menos {b} é igual a: {subtracao}")
print ("|                                                                       |")
print ("|----------------------------multiplicação------------------------------|")
print ("|                                                                       |")
print (f"o resultado da multiplicação de {a} vezes {b} é igual a: {multiplicacao}")
print ("|                                                                       |")
print ("|--------------------------------divisão--------------------------------|")
print ("|                                                                       |")
print (f"o resultado da divisão de {a} dividido por {b} é igual a: {divisao}")
```
